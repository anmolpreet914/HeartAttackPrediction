> While checking the data types for particular column, we find out that categorical column has the data type as 'O'. It can lead to decreased performance and increased memory usage compared to using more specific data types.

```python
# Iterate over columns
for column in df.columns:
    # Check if the dtype of the column is 'object'
    if df[column].dtype == 'object':
        # Convert the column to categorical dtype
        df[column] = df[column].astype('category')

# Check the data types after conversion
print(df.dtypes)
```

# Data Visualization

```python
# Count the occurrences of each risk level
risk_counts = df['Heart Attack Risk'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(risk_counts, labels=risk_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Heart Attack Risk Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Group data by continent and heart attack risk, and count occurrences
grouped = df.groupby(['Continent', 'Heart Attack Risk']).size().unstack(fill_value=0)

# Plot a pie chart for each continent
num_cont = len(grouped)
num_rows = (num_cont + 1) // 2  # Calculate number of rows for subplots
fig, axs = plt.subplots(num_rows, 2, figsize=(12, 4*num_rows))
fig.suptitle('Heart Attack Risk Distribution by Continent', fontsize=16)

for i, (continent, data) in enumerate(grouped.iterrows()):
    ax = axs[i // 2, i % 2]  # Adjust subplot index based on the number of continents
    ax.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140)
    ax.set_title(continent)

plt.tight_layout()
plt.show()
```

> Almost 34 to 37% of contribution in every continent can be seen of those who are more likely to develop the chances of Heart Attack.

```python
# Group by country and calculate the sum of '1's in 'Heart Attack Risk' column
risk_counts = df.groupby('Country')['Heart Attack Risk'].sum()

# Sort the counts in descending order and select the top 10 countries
top_10_countries = risk_counts.sort_values(ascending=False).head(10)

# Plot a bar chart
plt.figure(figsize=(10, 6))
top_10_countries.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries More Likely to Have Heart Attacks')
plt.xlabel('Country')
plt.ylabel('Number of Individuals at Risk')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
facet = sns.FacetGrid(df, hue="Heart Attack Risk",aspect=3)
facet.map(sns.kdeplot,"Age",fill= True)
facet.set(xlim=(0, df["Age"].max()))
facet.add_legend()
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: Heart Attack Risk by Sex
sns.countplot(data=df, x='Sex', hue='Heart Attack Risk', ax=axes[0, 0])
axes[0, 0].set_title('Heart Attack Risk by Sex')

# Plot 2: Age Distribution by Sex
sns.histplot(data=df, x='Age', hue='Sex', kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Age Distribution by Sex')

# Plot 3: Smoking Status by Sex
sns.countplot(data=df, x='Sex', hue='Smoking', ax=axes[1, 0])
axes[1, 0].set_title('Smoking Status by Sex')

# Plot 4: Exercise Hours Per Week by Sex
sns.boxplot(data=df, x='Sex', y='Exercise Hours Per Week', ax=axes[1, 1])
axes[1, 1].set_title('Exercise Hours Per Week by Sex')

plt.tight_layout()
plt.show()
```

> 1. Males demonstrate a greater likelihood of developing heart attacks compared to females, with the number of males at high risk of heart attacks doubling that of females.
> 2. Smoking is identified as a significant factor contributing to heart attacks, with all males being smokers. Despite this, approximately 2000 males are at risk of experiencing a heart attack.

```python
# Plot violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Smoking', y='Age', hue='Smoking', palette='viridis', legend=False)
plt.title('Relationship Between Smoking Status and Age')
plt.xlabel('Smoking Status (1 = Smoker, 0 = Non-smoker)')
plt.ylabel('Age')
plt.tight_layout()
plt.show()
```
